```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp

# Define the differential equation (same as before)
def dxdt(t, x):
    return np.exp(-x**3 + x**2 + 6*x) - 1

# Set simulation parameters
t_start = 0
t_end = 10
t_span = [t_start, t_end]

# Generate 10 initial conditions in the range [-3, 5]
initial_conditions = np.linspace(-3, 5, 10)

# Create the plot
plt.figure(figsize=(10, 6))

# Loop through each initial condition, solve, and plot
for x0_val in initial_conditions:
    sol = solve_ivp(dxdt, t_span, [x0_val], dense_output=True)
    t_eval = np.linspace(t_start, t_end, 200)
    x_sol = sol.sol(t_eval)
    plt.plot(t_eval, x_sol[0], label=f'x(0) = {x0_val:.2f}')

# Add horizontal lines for the fixed points to aid visualization
plt.axhline(y=3, color='g', linestyle='--', label='Stable Fixed Point (x=3)')
plt.axhline(y=-2, color='g', linestyle='--', label='Stable Fixed Point (x=-2)')
plt.axhline(y=0, color='r', linestyle='--', label='Unstable Fixed Point (x=0)')

# Formatting the plot
plt.title('Solutions for Various Initial Conditions')
plt.xlabel('Time (t)')
plt.ylabel('x(t)')
plt.ylim(-3.5, 5.5)
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.tight_layout()
plt.show()

```

```python
import numpy as np
import matplotlib.pyplot as plt

# 1. CORRECTLY NAMED DERIVATIVE FUNCTION
# This function defines the ODE: dx/dt = f(x)
def dxdt(x):
    return np.exp(-x**3 + x**2 + 6*x) - 1

# 2. CORRECT TIME INTERVAL AND PARAMETERS
dt = 0.1  # Using a slightly smaller dt for better stability
t_start = 0
t_end = 10
# Create the time vector
t = np.arange(t_start, t_end + dt, dt)

# 3. CORRECT RANGE FOR INITIAL CONDITIONS
# Generate 10 initial conditions in the specified range [-3, 5]
x0_range = np.linspace(-3, 5, 10)

# Create a matrix to hold all the solutions
# Each row will be a solution for a different initial condition
x_M = np.zeros((len(x0_range), len(t)))

# Fill the first column of our solution matrix with the initial conditions
x_M[:, 0] = x0_range

# --- Euler Method Loop (Vectorized) ---
# This loop calculates the solution for all initial conditions at once
for j in range(t.shape[0] - 1):
    x_M[:, j+1] = x_M[:, j] + dt * dxdt(x_M[:, j])

# --- Plotting the Results ---
plt.figure(figsize=(10, 6))
# The .T transposes the matrix so we can plot each solution (row) easily
plt.plot(t, x_M.T)

# Add horizontal lines for the fixed points to compare
plt.axhline(y=3, color='g', linestyle='--', label='Stable Fixed Point (x=3)')
plt.axhline(y=-2, color='g', linestyle='--', label='Stable Fixed Point (x=-2)')
plt.axhline(y=0, color='r', linestyle='--', label='Unstable Fixed Point (x=0)')

# Formatting the plot
plt.title("Solutions using Euler's Method")
plt.xlabel('Time (t)')
plt.ylabel('x(t)')
plt.ylim(-3.5, 5.5) # Set y-axis limits to match the previous plot
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the range for r
r_neg = np.linspace(-2, 0, 400)
r_pos = np.linspace(4, 5, 400)
r_full = np.linspace(-2, 5, 400)

# Fixed point x* = 0
x1_stable = np.zeros_like(r_neg)
x1_unstable = np.zeros_like(np.linspace(0, 5, 400))

# Fixed points x* = (-r +/- sqrt(r^2 - 4r)) / 2
# For r <= 0 (both are unstable)
x2_unstable_neg = (-r_neg + np.sqrt(r_neg**2 - 4*r_neg)) / 2
x3_unstable_neg = (-r_neg - np.sqrt(r_neg**2 - 4*r_neg)) / 2

# For r >= 4 (one stable, one unstable)
x2_stable_pos = (-r_pos + np.sqrt(r_pos**2 - 4*r_pos)) / 2
x3_unstable_pos = (-r_pos - np.sqrt(r_pos**2 - 4*r_pos)) / 2


# Plotting
plt.figure(figsize=(10, 6))

# Plot stable branches (solid lines)
plt.plot(r_neg, x1_stable, 'b-', label='Stable')
plt.plot(r_pos, x2_stable_pos, 'b-')

# Plot unstable branches (dashed lines)
plt.plot(np.linspace(0, 5, 400), x1_unstable, 'r--', label='Unstable')
plt.plot(r_neg, x2_unstable_neg, 'r--')
plt.plot(r_neg, x3_unstable_neg, 'r--')
plt.plot(r_pos, x3_unstable_pos, 'r--')

# Formatting
plt.title('Bifurcation Diagram for $\dot{x} = x^3 + rx^2 + rx$')
plt.xlabel('Parameter r')
plt.ylabel('Fixed points $x^*$')
plt.grid(True)
plt.ylim(-5, 3)
plt.xlim(-2, 5)
plt.legend()
plt.show()
```

```python
# simulate_scalar_ode.py
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp

# --- define the ODE ---
def f(t, x):
    # x may be array-like; ensure scalar
    xx = np.asarray(x)
    return np.exp(-xx**3 + xx**2 + 6*xx) - 1

# fixed points (computed analytically)
fixed_points = np.array([-2.0, 0.0, 3.0])

# time span and evaluation times
t0, tf = 0.0, 30.0
t_eval = np.linspace(t0, tf, 1000)

# initial conditions: at least 10 different in range [-3, 5]
x0_list = np.linspace(-3.0, 5.0, 12)  # 12 initial conditions (>=10)

# prepare the plot
plt.figure(figsize=(9,5))
for x0 in x0_list:
    sol = solve_ivp(f, (t0, tf), [x0], t_eval=t_eval, vectorized=False, rtol=1e-8, atol=1e-10)
    plt.plot(sol.t, sol.y[0], lw=1.5, label=f"x0={x0:.2f}")

# mark fixed points as horizontal lines
for xp in fixed_points:
    plt.hlines(xp, t0, tf, linestyles='--', linewidth=2, alpha=0.8)
    plt.text(tf*0.99, xp, f"  x*={xp}", va='center', ha='right')

plt.xlabel("t")
plt.ylabel("x(t)")
plt.title(r"Trajectories of $\dot x = e^{-x^3 + x^2 + 6x} - 1$ for various $x(0)$")
plt.xlim(t0, tf)
plt.ylim(-3.5, 4.5)
plt.grid(True)
# optional: show legend (can be busy if many curves)
# plt.legend(loc='upper right', bbox_to_anchor=(1.35, 1))
plt.tight_layout()
plt.show()

# --- Optional: plot phase portrait (x vs f(x)) and indicate flow direction ---
xs = np.linspace(-4.0, 5.0, 400)
fxs = np.exp(-xs**3 + xs**2 + 6*xs) - 1
plt.figure(figsize=(8,3.5))
plt.plot(xs, fxs, lw=2)
plt.axhline(0, color='k', linewidth=0.6)
plt.scatter(fixed_points, np.zeros_like(fixed_points), zorder=5)
for xp in fixed_points:
    plt.text(xp, 0.02, f"{xp}", ha='center')
plt.xlabel("x")
plt.ylabel("f(x) = $\dot x$")
plt.title("Phase portrait: f(x) vs x (zeros are fixed points)")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp

def dxdt(x):
    return np.exp(-x**3 + x**2 + 6*x) - 1

def dxdt_IVP(t, x):
    return np.exp(-x**3 + x**2 + 6*x) - 1

# Common time setup
dt = 0.01
t_start, t_end = 0, 10
t = np.arange(t_start, t_end + dt, dt)
x0 = 1.0  # initial condition

# --- Euler Method ---
x_euler = np.zeros_like(t)
x_euler[0] = x0
for j in range(len(t) - 1):
    x_euler[j+1] = x_euler[j] + dt * dxdt(x_euler[j])

# --- RK4 Method ---
x_rk = np.zeros_like(t)
x_rk[0] = x0
for j in range(len(t) - 1):
    k1 = dxdt(x_rk[j])
    k2 = dxdt(x_rk[j] + 0.5 * dt * k1)
    k3 = dxdt(x_rk[j] + 0.5 * dt * k2)
    k4 = dxdt(x_rk[j] + dt * k3)
    x_rk[j+1] = x_rk[j] + dt / 6 * (k1 + 2*k2 + 2*k3 + k4)

# --- solve_ivp (RK45 default) ---
sol = solve_ivp(dxdt_IVP, [t_start, t_end], [x0], t_eval=t, dense_output=True)
x_ivp = sol.y[0]

# --- Plot ---
plt.figure(figsize=(8, 4))
plt.plot(t, x_euler, label='Euler', alpha=0.7)
plt.plot(t, x_rk, '--', label='RK4', alpha=0.8)
plt.plot(t, x_ivp, '-.', label='solve_ivp (RK45)', alpha=0.9)
plt.xlabel("t")
plt.ylabel("x(t)")
plt.title("Comparison of Integration Methods")
plt.legend()
plt.grid(True)
plt.show()
```
