##### Q1 d Rendering Phase Potrait

```python
def f(x,y):
    return [ (1-x-y**2), (x*y + 3*y)]
```

```python
# Define range of x and y values
x_range = np.linspace(-10,10,100)
y_range = np.linspace(-10,10,100)

# Make meshgrid
xm, ym = np.meshgrid(x_range, y_range)

# Evaluate the velocity vector for each set of x and y values
dxm, dym = f(xm,ym)

# Plot
plt.figure(figsize=(6,6))
plt.streamplot(xm, ym, dxm, dym, color=colours[7])

# Add in nullclines
plt.plot(x_range,-np.sqrt(1-x_range), color=colours[3])
plt.plot(x_range, np.sqrt(1-x_range), color=colours[3])

plt.plot(x_range, np.zeros(100), color=colours[0])
plt.plot(-3*np.ones(100), y_range, color=colours[0])



plt.xlabel('x')
plt.ylabel('y')

plt.grid()
```

```python
#Checking on Eigen values

def jacobian(x,y):
    return np.array([[-1, -2*y],
                     [y, x+3]])

print(np.linalg.eig(jacobian(1,0))
,np.linalg.eig(jacobian(-3,2))
,np.linalg.eig(jacobian(-3,-2)))
```

##### Q1.e Simulating system and solution given Initial Condition

```python
def f_ivp(t, u):
    x0 = u[0]
    y0 = u[1]
    return [1 - x0 - y0**2, x0*y0 + 3*y0]
```

```python
tfinal = 10
plusx0 = 1
plusy0 = 1
minusx0 = -1
minusy0 = -1
```

```python
sol = solve_ivp(f_ivp,[0, tfinal], [plusx0, plusy0], dense_output=True)
sol2 = solve_ivp(f_ivp,[0, tfinal], [minusx0, minusy0], dense_output=True)
t = np.linspace(0,tfinal,100)
x,y = sol.sol(t)
x2,y2 = sol2.sol(t)
```

```python
plt.plot(t, x, label='Species 1')
plt.plot(t, y, label='Species 2')

plt.title('Simulated solution w.r.t IC 1')
plt.xlabel('Magnitude')
plt.ylabel('Time')
```

```python
plt.plot(t, x2, label='Species 1')
plt.plot(t, y2, label='Species 2')

plt.title('Simulated solution w.r.t IC 2')
plt.xlabel('Magnitude')
plt.ylabel('Time')
```

#### Q2 b  Reference on Ch8 coding exercise solution Q3a

```python
def Q2(t,u,mu):
    x = u[0]
    y = u[1]
    return [ mu*x - y - x**3, x + mu*y - y**3]

# Defining the system
```

```python
tfinal = 50
x0_1 = [0,1]
x0_2 = [0.5,0]

# Define range of x and y values
x_range2 = np.linspace(-1.5,1.5,100)
y_range2 = np.linspace(-1.5,1.5,100)

# Make meshgrid
xm, ym = np.meshgrid(x_range2,y_range2)

#Setting up range of parameter 
```

##### Plot with nullclines

```python
# Create figure
fig = plt.figure(figsize=(12,4))

i = 1
for mu in [-1, 1, 4]:
    
    # Simulate the system
    sol1 = solve_ivp(Q2, [0, tfinal], x0_1, args=(mu,),dense_output=True)
    sol2 = solve_ivp(Q2, [0, tfinal], x0_2, args=(mu,),dense_output=True)
    t = np.linspace(0, tfinal, 1000)
    x1 = sol1.sol(t)[0]
    y1 = sol1.sol(t)[1]
    x2 = sol2.sol(t)[0]
    y2 = sol2.sol(t)[1]
    
    # Evaluate the velocity vector for each set of x and y values
    dxm, dym = Q2(t,[xm,ym],mu)

    # Set up subplot
    plt.subplot(1,3,i)
    
    # Plot vector field
    plt.streamplot(xm, ym, dxm, dym, color=colours[7])
    
# Add nullclines
    plt.plot(x_range2, mu*x_range2 -x_range2**3, color=colours[2], label='x nulcline')
    

    plt.plot(y_range2**3 - mu*x_range2, y_range2, color=colours[1], label='y nulcline')

    plt.axis([-1.5,1.5,-1.5,1.5])
    plt.xlabel('$x$',size= 22)
    plt.ylabel('$y$',size= 22)
    plt.grid()
    
    i += 1

    plt.plot(x_range2,mu*x_range2 - x_range2**3,color=colours[3])
    plt.plot(y_range2, y_range2**3 - mu*y_range2,color=colours[0])
    plt.plot(np.zeros(100),y_range2,color=colours[0])
   
    plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
   
```

##### Plot with solution

```python
# Create figure
fig = plt.figure(figsize=(12,4))

i = 1
for mu in [-1, 1, 4]:
    
    # Simulate the system
    sol1 = solve_ivp(Q2, [0, tfinal], x0_1, args=(mu,),dense_output=True)
    sol2 = solve_ivp(Q2, [0, tfinal], x0_2, args=(mu,),dense_output=True)
    t = np.linspace(0, tfinal, 1000)
    x1 = sol1.sol(t)[0]
    y1 = sol1.sol(t)[1]
    x2 = sol2.sol(t)[0]
    y2 = sol2.sol(t)[1]
    
    # Evaluate the velocity vector for each set of x and y values
    dxm, dym = Q2(t,[xm,ym],mu)

    # Set up subplot
    plt.subplot(1,3,i)
    
    # Plot vector field
    plt.streamplot(xm, ym, dxm, dym, color=colours[7])
    
    # Plot simulated solutions
    plt.plot(x1,y1,color='b')
    plt.plot(x2,y2,color='r')

    plt.axis([-1.5,1.5,-1.5,1.5])
    plt.xlabel('$x$',size= 22)
    plt.ylabel('$y$',size= 22)
    plt.grid()
    
    i += 1

    plt.plot(x_range2,mu*x_range2 - x_range2**3,color=colours[3])
    plt.plot(y_range2, y_range2**3 - mu*y_range2,color=colours[0])
    plt.plot(np.zeros(100),y_range2,color=colours[0])
```
